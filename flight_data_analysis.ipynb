```python
##6 which airline has the highest capacity of seats?
print( df_pl.columns, df_fl.columns, sep='\n')
```

```python
carrier_tailno = df_fl[['carrier','tailnum']].drop_duplicates()
carrier_tailno
```

```python
pd.merge(carrier_tailno, df_pl[['tailnum','seats']] ,how = 'left', on='tailnum')
```

```python
aeroplane = pd.merge(carrier_tailno, df_pl[['tailnum','seats']] ,how = 'left', on='tailnum').groupby('carrier').agg(total_seats = ('seats', 'sum' )).sort_values(by='total_seats', ascending=False)
aeroplane
```

```python
##7 which airplane model is the highest in use and from which manufacturer?

aero_use = df_fl.groupby('tailnum').agg(counts  = ('tailnum', 'count')).sort_values('counts',ascending=False).reset_index()
aero_use
```

```python
res7 = pd.merge( df_pl[['tailnum','manufacturer','model']], aero_use,how='left',on='tailnum').groupby(['model','manufacturer'])\
                       .agg(total_flight = ('counts','sum')).sort_values('total_flight',ascending=False)

res7 #airbus
```
